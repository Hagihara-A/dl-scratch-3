```python
from dezero import Variable
import dezero.functions as F
import dezero.layers as L
from dezero.optimizers import MomentumSGD as opt
import numpy as np
from dezero.models import MLP
np.random.seed(0)
N, W = 100, 1
x = np.random.rand(N, W)
y = np.sin(2 * np.pi * x) + np.random.rand(N, W)


model = MLP((1000, W))
optimizer = opt(0.002).setup(model)
iters = 5000

for i in range(iters):
    y_pred = model(x)
    loss = F.mean_squared_error(y, y_pred)
    model.clear_grads()
    loss.backward()
    optimizer.update()
    if (i%250==0):
        print(f'{i} times iterated')
    if (i % 1000 == 0):
        print("loss: ",loss.data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.close()

plt.scatter(x, y)
plt.scatter(x, model(x).data)
plt.show()
```
